## Analyzing data

```python
df.corr()
```

```python
#creating heatmap for correlation between various continuous valued columns
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True)
```

<b>From above heatmap we can conclude that sales and profit are highly correlated.
Second highest correlation is between sales and quantity.
Discount and profit are negatively correlated, i.e if discount increases, profit decreases.</b>

```python
sns.countplot('Ship Mode', data=df)
```

<b>Most of the shipping is done through standard class and least is done on same day</b>

```python
sns.countplot('Segment', data=df)
```

```python
sns.countplot('Region', data=df)
```

<b>Major portion of customers are from West, then east, then central and least from South</b>

```python
sns.countplot('Category', data=df)
```

<b>They supply office goods mostly and less no. of technology and furniture</b>

<b><i>Pairplots are a really simple way to visualize relationships between each variable. It produces a matrix of relationships between each variable in the data for an instant examination of our data.</i></b>

```python
sns.pairplot(df)
```

```python
pd.crosstab(df['Ship Mode'],df['Category']).plot(kind='bar',stacked=False)
```

```python
pd.crosstab(df['Ship Mode'],df['Region']).plot(kind='bar')
```

```python
pd.crosstab(df['Segment'],df['Region']).plot(kind='bar')
```

```python
category_profit= df.groupby(['Category'])['Profit'].sum().nlargest()
category_profit
```

```python
plt.figure(figsize=(5,5))
labels=df['Category'].unique()
plt.pie(df.groupby(['Category'])['Profit'].sum(),autopct='%1.1f%%',labels=labels)

plt.title('Distribution of categorywise profit',size=13)

plt.show()
```

```python
category_quantity = df.groupby(['Category'])['Quantity'].sum().nlargest()
category_quantity
```

```python
plt.figure(figsize=(5,5))
labels=df['Category'].unique()
plt.pie(df.groupby(['Category'])['Quantity'].sum(),autopct='%1.1f%%',labels=labels)

plt.title('Distribution of products quantity',size=13)

plt.show()
```

<b>Maximum profit comes from technology but quantity of ffice supplies is more</b>

```python
region_profit= df.groupby(['Region'])['Profit'].sum().nlargest()
region_profit
```

<b>Maximum profit comes from west region</b>

```python
states_profit= df.groupby(['State'])['Profit'].sum().nlargest(10)
states_profit
```

<b>Maximum profit comes from California</b>

```python
city_profit= df.groupby(['City'])['Profit'].sum().nlargest(10)
city_profit
```

<b>Maximum profit comes from New York City</b>

```python
category_discount= df.groupby(['Category'])['Discount'].sum().nlargest()
category_discount
```

<b>Discount decreases profit and maximum discount is provided on funiture items</b>

```python
sub_category_profit= df.groupby(['Sub-Category'])['Profit'].sum().nlargest()
sub_category_profit
```

<b>Maximum profit comes from copiers</b>

```python
df.groupby("Category")[['Profit', 'Sales']].sum().plot.bar(color = ["green", "red"])
plt.ylabel("Profit /loss and Sales")
plt.show()
```

<b>Sales and profit is highest for Technology category</b>

```python
df.groupby("Sub-Category")[['Profit', 'Sales']].sum().plot.bar(color = ["green", "red"], figsize=(17,7))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

<b>Sales of chair is highest but maximum profit comes from copiers</b>
